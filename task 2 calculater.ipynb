```python
def add(numbers):
    return sum(numbers)


def subtract(numbers):
    result = numbers[0]
    for number in numbers[1:]:
        result = result - number
    return result


def multiply(numbers):
    result = numbers[0]
    for number in numbers[1:]:
        result = result * number
    return result


def divide(numbers):
    result = numbers[0]
    for number in numbers[1:]:
        result = result / number
    return result


def even_odd(number):
    if number % 2 == 0:
        print("Even")
    else:
        print("Odd")


def factorial(number):
    fact = 1
    for i in range(1, number + 1):
        fact = fact * i
    return fact


# Program start
flag = True
while flag:
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Even or Odd")
    print("6. Factorial")

    userInput = input("Choose an operation: ")

    if userInput == "1" or userInput == "2" or userInput == "3" or userInput == "4":
        numbers_count = int(input("How many numbers ? : "))
        numbers = []
        for i in range(numbers_count):
            numbers.append(float(input("Enter number : ")))
        if userInput == "1":
            print(add(numbers))
        elif userInput == "2":
            print(subtract(numbers))
        elif userInput == "3":
            print(multiply(numbers))
        elif userInput == "4":
            print(divide(numbers))

    elif userInput == "5" or userInput == "6":
        number = int(input("Enter number : "))
        if userInput == "5":
            even_odd(number)
        else:
            print(factorial(number))

    # ask user for another operation
    choice = input("Another operation ? 'yes', 'no'")
    if choice == "no":
        print("Done")
        flag = False
```
